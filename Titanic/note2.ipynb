```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
```

```python
train_df = pd.read_csv('train.csv')
test_df = pd.read_csv('test.csv')
```

```python
train_df.drop(['PassengerId','Name', 'Ticket','Cabin','Embarked'], axis=1, inplace=True)
```

```python
train_df
```

```python
train_df['m'] = train_df['Sex'] == 'male'
```

```python
train_df
```

```python
train_df['f'] = train_df['m'] == False
```

```python
train_df.drop(['Sex'], axis=1, inplace=True)
```

```python
mean = train_df['Age'].mean(skipna=True)
train_df['Age'].fillna(value=mean, inplace=True)
```

```python
train_df
```

```python
clf = RandomForestClassifier(n_jobs=2, random_state=0)
y = train_df['Survived']
x = train_df.iloc[:,2:]
clf.fit(x,y)
```

```python
test_df.drop(['Name','Ticket','Cabin','Embarked'], axis=1, inplace=True)
test_df['m'] = test_df['Sex'] == 'male'
test_df['f'] = test_df['m'] == False
mean = test_df['Age'].mean(skipna=True)
test_df['Age'].fillna(value=mean, inplace=True)
mean = test_df['Fare'].mean(skipna=True)
test_df['Fare'].fillna(value=mean, inplace=True)
```

```python
result = clf.predict(test_df.iloc[:,2:])
final = list(zip(test_df['PassengerId'], result))
final_prediction = pd.DataFrame(final, columns=['PassengerId', 'Survived'])
final_prediction.to_csv('final_prediction.csv', index=False)
```
